```python
#simple linear regression project
#importing necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#sample data
height = np.array([60, 61, 62, 63, 64, 65, 66, 67, 68, 69]) #in inches
weight = np.array([110, 120, 130, 140, 150, 160, 170, 180, 190, 200]) #in pounds

#reshaping the data
height = height.reshape(-1, 1)

#Creating linear regression model
model = LinearRegression()

#Training model on data
model.fit(height, weight)

#Make predictions for new height
new_height = np.array([70]).reshape(-1, 1)
predicted_weight = model.predict(new_height)
print(f"predicted Weight: {predicted_weight[0]:.2f} pounds")

#plot the data and regression line
plt.scatter(height, weight, color="blue", label="Data Points")
plt.plot(height, model.predict(height), color="red", label="Regression Line")
plt.legend()
plt.xlabel("Height (in inches)")
plt.ylabel("Weight (in pounds)")
plt.title("Linear Regression: Height vs Weight")
plt.show

```
