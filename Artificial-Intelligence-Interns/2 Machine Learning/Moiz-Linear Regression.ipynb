```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic data
np.random.seed(0)
height = np.random.normal(160, 10, 100)  
weight = 50 + 0.6 * height + np.random.normal(0, 5, 100)  

# Create and fit a Linear Regression model
model = LinearRegression()
height = height.reshape(-1, 1)  
model.fit(height, weight)

# Make predictions
height_test = np.linspace(140, 180, 100).reshape(-1, 1)
weight_pred = model.predict(height_test)

# Calculating the model's performance
mse = mean_squared_error(weight, model.predict(height))
r2 = r2_score(weight, model.predict(height))
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared (R2) Score: {r2:.2f}')

# Visualize the LR model
plt.scatter(height, weight, color='blue', label='Actual Data')
plt.plot(height_test, weight_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.title('Linear Regression: Weight Prediction')
plt.show()
```
